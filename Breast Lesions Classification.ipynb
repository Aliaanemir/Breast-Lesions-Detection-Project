```python
import pandas as pd
import numpy as np
import cv2 
import csv
import xml.etree.ElementTree as et
import pydicom
```

```python
xml_path = "INbreast_20586908/20586908.xml"

tree = et.parse(xml_path)
root = tree.getroot()

mammogram_data = open("20586908.csv", 'w')
        
```

```python
maxdepth = 0
def depth(elem, level):
    global maxdepth
    if (level == maxdepth):
        maxdepth += 1
    print('layer', maxdepth)    
    for child in elem:
        depth(child, level + 1)
        print(child.text)
        
depth(root, -1)
print(maxdepth)
```

```python
def parse_children(root, maxdepth):
    for _ in range(maxdepth):
        parse_children(root)
    
for _ in range(maxdepth):
    parse_children(root)
```
